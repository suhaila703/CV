```python
import cv2
import numpy as np
from google.colab.patches import cv2_imshow
```

```python
from google.colab import drive
drive.mount('/content/drive')
img_path = '/content/drive/MyDrive/images/persimmons.webp'  # ðŸ‘ˆ change this
img = cv2.imread(img_path)

# Show the original
cv2_imshow(img)
```

```python
# === 2ï¸âƒ£ Convert to grayscale ===
gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

# === 3ï¸âƒ£ Invert the grayscale image ===
inverted = cv2.bitwise_not(gray)

# === 4ï¸âƒ£ Apply Gaussian Blur to the inverted image ===
blurred = cv2.GaussianBlur(inverted, (21, 21), 0)   # kernel size controls smoothness

# === 5ï¸âƒ£ Display results ===
cv2_imshow(gray)
cv2_imshow(inverted)
cv2_imshow(blurred)
```

```python
# Formula: sketch = gray / (255 - blurred) * scale
sketch = cv2.divide(gray, 255 - blurred, scale=256)

# === 6ï¸âƒ£ Display the result ===
cv2_imshow(sketch)
```
